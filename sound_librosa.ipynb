```python
import numpy as np, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
```

```python
x, sr = librosa.load('a.wav')
ipd.Audio(x, rate=sr)
```

```python
librosa.display.waveplot(x, sr)
```

```python
plt.stem(x[40000:40100])
plt.ylabel('pressure scaled to values between -1 and 1')
plt.xlabel('samples taken every 1/sr of a second')
```

```python
hop_length = 32
S = librosa.stft(x, n_fft=round(0.005*sr), hop_length=hop_length)

plt.figure(figsize=(15, 5))
librosa.display.specshow(S, sr=sr, x_axis='time', y_axis='fft', fmax = 5000)

plt.colorbar()
```
